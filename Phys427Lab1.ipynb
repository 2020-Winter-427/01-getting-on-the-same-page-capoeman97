```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.normal(loc = 0, scale = 1, size = 1000)
y = np.random.normal(loc = 0, scale = 1, size = 1000)

#genarates 2 arrays of random numbers from a normal distribution

plt.hist(y, bins = 10) #plots histogram with y array as data input
plt.title("Normal Distribution Histogram")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

plt.scatter(x, y, c = 'r', s = 5) #scatter plot with x and y as arrays, color red, and scale of dots at 5
plt.title("Normal Distribution Scatter Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x1 = np.random.exponential(scale = 1, size = 1000)
y1 = np.random.exponential(scale = 1, size = 1000)

#genarates 2 arrays of random data from an exponential distribution

plt.hist(y1, bins = 10)
plt.title("Exponential Distribution Histogram")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

plt.scatter(x1, y1, c = 'r', s = 5)
plt.title("Exponential Distribution Scatter Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
